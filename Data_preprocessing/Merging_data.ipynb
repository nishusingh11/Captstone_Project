```python
read_location = '/content/drive/MyDrive/BDA600_FinalProject/data/Individual_files/'
write_location = '/content/drive/MyDrive/BDA600_FinalProject/data/'

files = str(sp.check_output(['ls', read_location])).split('\\n')[:-1]
files[0] = files[0][2:]

print(files)
```

```python
for i in range(len(files)):
    globals()[f"df_{i}"] = pd.read_csv(read_location+files[i])
```

```python
frames = []
frames.extend(value for name, value in locals().items() if name.startswith('df_'))

data = pd.concat(frames) 
print(len(data))
data.head(2)
```

```python
print(min(data['time']), max(data['time']))
```

```python
data.to_csv(write_location+'Data_2000-23.csv', index=False)
```
